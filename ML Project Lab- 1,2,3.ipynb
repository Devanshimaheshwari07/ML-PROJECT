```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("student_prediction.csv")

# Select important numerical predictor variables
important_features = ['STUDY_HRS', 'CUML_GPA', 'ATTEND']
num_df = df[important_features]


# Descriptive Statistics
mean = num_df.mean()
median = num_df.median()
mode = num_df.mode().iloc[0]
std_dev = num_df.std()
minimum = num_df.min()
maximum = num_df.max()
total_sum = num_df.sum()


# Quartiles & Percentiles
q1 = num_df.quantile(0.25)
q2 = num_df.quantile(0.50)
q3 = num_df.quantile(0.75)
p90 = num_df.quantile(0.90)

# Correlation & Covariance
correlation = num_df.corr()
covariance = num_df.cov()

# Combined results
stats_df = pd.DataFrame({
    "Mean": mean,
    "Median": median,
    "Mode": mode,
    "Std Deviation": std_dev,
    "Min": minimum,
    "Max": maximum,
    "Sum": total_sum,
    "Q1 (25%)": q1,
    "Q2 (50%)": q2,
    "Q3 (75%)": q3,
    "90th Percentile": p90
})


# Display Results
print("Descriptive Statistical Summary (Important Variables):\n")
print(stats_df)

print("\nCorrelation Matrix:\n")
print(correlation)

print("\nCovariance Matrix:\n")
print(covariance)
```
